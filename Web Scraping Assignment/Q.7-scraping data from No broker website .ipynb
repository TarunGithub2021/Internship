```python
house=[]

for i in soup.find_all('h2',class_="heading-6 font-semi-bold nb__1AShY"):
    house.append(i.text.split('In')[0:-1])
    
house
```

```python
area=[]

for i in soup.find_all('h2',class_="heading-6 font-semi-bold nb__1AShY"):
    area.append(i.text.split('In')[-1])
    
area
```
